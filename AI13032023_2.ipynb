```python
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show

```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x='Salary', y='YearsExperience', figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model #mo hinh tuyen tinh
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']), y= data['YearsExperience'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(pd.DataFrame(data['Salary']),data['YearsExperience'] ) #do chinh xac bao nhieu %
```

```python
data.plot(kind = 'scatter', x='Salary', y='YearsExperience', figsize =(9,9), color = 'black', xlim=(0,7))
y_pred = model.predict(X = pd.DataFrame(data['Salary']))
plt.plot(data['Salary'], y_pred, color = 'blue')
#day la duong hoi quy
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2]).T
model.fit(X,y = data['YearsExperience'])
print('Intercept cuar mo hinh la:' ,model.intercept_)
print('Coeffient cua mo hinh la:' ,model.coef_)
print('Do chinh xac cua mo hinh:' ,model.score(X, y = data['YearsExperience']))
```

```python
import numpy as np
data.plot(kind = 'scatter', x='Salary', y='YearsExperience', figsize =(9,9), color = 'red', xlim=(0,7))
range = np.arange(1.5,15000,0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```
