```python
import pandas as pd
url = 'Advertising.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='TV', y='sales', figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model #mo hinh tuyen tinh
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['TV']), y= data['sales'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(pd.DataFrame(data['TV']),data['sales'] )
```

```python
data.plot(kind = 'scatter', x='TV', y='sales', figsize =(9,9), color = 'black', xlim=(0,300))
y_pred = model.predict(X = pd.DataFrame(data['TV']))
plt.plot(data['TV'], y_pred, color = 'blue')
```
