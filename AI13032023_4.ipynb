```python
import pandas as pd
url = '50_Startups.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show


```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='R&D Spend', y='Profit', figsize = (9,9), color = 'black')

from sklearn import linear_model #mo hinh tuyen tinh
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['R&D Spend']), y= data['Profit'])
print(model.intercept_)
print(model.coef_)
model.score(pd.DataFrame(data['R&D Spend']),data['Profit'] ) #do chinh xac bao nhieu %
```

```python
data.plot(kind = 'scatter', x='R&D Spend', y='Profit', figsize =(9,9), color = 'black', xlim=(0,150000))
y_pred = model.predict(X = pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'], y_pred, color = 'blue')
#day la duong hoi quy
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**2]).T
model.fit(X,y = data['Profit'])
print('Intercept cuar mo hinh la:' ,model.intercept_)
print('Coeffient cua mo hinh la:' ,model.coef_)
print('Do chinh xac cua mo hinh:' ,model.score(X, y = data['Profit']))
```

```python
import numpy as np
data.plot(kind = 'scatter', x='R&D Spend', y='Profit', figsize =(9,9), color = 'black', xlim=(0,150000))
range = np.arange(0,150000,1000)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**2,data['R&D Spend']**3 ] ).T
model.fit(X,y = data['Profit'])
print('Intercept cuar mo hinh la:' ,model.intercept_)
print('Coeffient cua mo hinh la:' ,model.coef_)
print('Do chinh xac cua mo hinh:' ,model.score(X, y = data['Profit']))
```
