```python
from keras.datasets import mnist #khai báo tập dữ liệu keras trong mnits
(train_images,train_labels), (test_images,test_labels) = mnist.load_data()
```

```python
from keras import models
from keras.layers import Dense

from keras.layers.attention.multi_head_attention import activation
model = models.Sequential()
model.add(Dense(512, activation = 'relu', input_shape =(28*28,), name='layer_1'))
model.add(Dense(10, activation ='softmax', name='layer_out'))
model.summary()
```

```python
train_images = train_images.reshape((60000, 28*28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28*28))  # sửa lại thành biến test_images
test_images = test_images.astype('float32') / 255
```

```python
from keras.utils import to_categorical
train_labels = to_categorical(train_labels,10)
test_labels = to_categorical(test_labels,10)
print(train_labels.shape)
print(test_labels.shape)
```

```python
## data -> model -> compile -> train -> Evaluation 
```

```python
#compile
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])  #chon bo toi uu, danh gia do mat mat, do chinh xac  
#train
model.fit(train_images, train_labels, epochs = 10, batch_size = 128 ) # epochs: so lan hoc, bath_size: so du lieu moi lan hoc, boc ngau nhien 128
#evaluate, do do chinh xac
test_loss , test_acc = model.evaluate(test_images, test_labels)
print('Do mat mat :', test_loss)
print('Do chinh xac :', test_acc)
```
